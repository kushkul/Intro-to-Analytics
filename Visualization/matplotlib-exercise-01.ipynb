```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
mydata = pd.read_csv('company_sales_data.csv')
```

```python
mydata.head(2)
```

```python
plt.plot(mydata['month_number'], mydata['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
```

```python
plt.plot(mydata['month_number'], mydata['total_profit'], '--r', label = 'Profit per month', linewidth=3, marker='o')
#plt.scatter(mydata['month_number'], mydata['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc = 'lower right')
```

```python
#Q3
columns = mydata.columns[1:-2]
for i in columns:
    plt.plot(mydata['month_number'], mydata[i], label = 'Plot for %s'%(i), marker = 'o')
plt.legend()
plt.show()
```

```python
#Q4
plt.scatter(mydata['month_number'], mydata['toothpaste'])
```

```python
#Q5
monthlist = mydata['month_number'].tolist()
plt.bar([a-0.25 for a in monthlist], mydata['facecream'], width = 0.25, label = 'Facecream data', align = 'edge')
plt.bar([a+0.25 for a in monthlist], mydata['facewash'], width = -0.25, label = 'Fashwash data', align = 'edge')
plt.show()
```
